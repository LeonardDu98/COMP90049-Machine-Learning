```python
# function load csvdata
import pandas as pd

def load_csvdata(csvFilename):
    data = pd.read_csv(open(csvFilename))
    return data
```

```python
vo = open("vocab.txt", "r")
vocab_dic = {}
for i in vo.readlines():
    feature, index = i.replace("\n", "").split("\t") #re
    vocab_dic[index] = feature
```

```python
import re
def get_train_vector(train):
    result = []
    for i in train:
        temp = re.findall(r'[(](.*?)[)]', i)
        oneLine = [0 for _ in range(len(vocab_dic))]
        for j in temp:
            index, count = j.split(",")
            oneLine[int(index)] = float(count)
        result.append(oneLine)
    return result
```

```python
train_count_vector = get_train_vector(train_count)
train_tfidf_vector = get_train_vector(train_tfidf)

dev_count_vector = get_train_vector(dev_count)
dev_tfidf_vector = get_train_vector(dev_tfidf)

test_count_vector = get_train_vector(test_count)
test_tfidf_vector = get_train_vector(test_tfidf)
```
